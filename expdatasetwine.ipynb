```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/winequality-red.csv', sep=';')
display(df.head())
```

```python
df2 = pd.read_csv('/content/winequality-white.csv', sep=';')
display(df.head())
```

```python
combined_df = pd.concat([df, df2])
display(combined_df.head())
display(combined_df.tail())
```

```python
df.columns #nomes das colunas
```

```python
df.info() # qnt colunas e entradas
```

```python
pandas_mean = combined_df['alcohol'].mean() #media com pandas e numpy
numpy_mean = np.mean(combined_df['alcohol'])
print(f"Média do teor alcoólico (pandas): {pandas_mean}")
print(f"Média do teor alcoólico (numpy): {numpy_mean}")
print(f"Os resultados são iguais? {np.isclose(pandas_mean, numpy_mean)}")
```

```python
print("Verificando valores NaN no dataset:") #valores nao numericos
display(combined_df.isnull().sum())

print("\nCalculando a média do teor alcoólico (pandas, ignorando NaNs):")
pandas_mean_no_nan = combined_df['alcohol'].dropna().mean()
print(f"Média do teor alcoólico (pandas, sem NaNs): {pandas_mean_no_nan}")

print("\nCalculando a média do teor alcoólico (numpy, ignorando NaNs):")
numpy_mean_no_nan = np.nanmean(combined_df['alcohol'])
print(f"Média do teor alcoólico (numpy, com np.nanmean): {numpy_mean_no_nan}")

print(f"\nOs resultados de ambas as abordagens são iguais? {np.isclose(pandas_mean_no_nan, numpy_mean_no_nan)}")
```

```python
plt.figure(figsize=(10, 6)) # grafico teor alcolico por frequencia
sns.histplot(combined_df['alcohol'], bins=20, kde=True)
plt.title('Distribuição do Teor Alcoólico')
plt.xlabel('Teor Alcoólico')
plt.ylabel('Frequência')
```

```python
plt.figure(figsize=(10, 6)) # grafico teor alcolico por qualidade
sns.barplot(x='quality', y='alcohol', data=combined_df)
plt.title('Distribuição do Teor Alcoólico por Qualidade')
plt.xlabel('Qualidade')
plt.ylabel('Teor Alcoólico')
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=combined_df['pH'])
plt.title('Boxplot do pH')
plt.xlabel('pH')
plt.show()
```

```python
quality_distribution = combined_df['quality'].value_counts()
display(quality_distribution)
```

```python
correlation_matrix = combined_df.corr() # matriz de correlação

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Wine Quality Dataset')
plt.show()
```

```python
quality_correlation = correlation_matrix['quality'].sort_values(ascending=False)
print("Correlation with 'quality':")
display(quality_correlation)
```

```python
print("\nVariables with the strongest impact on 'quality' (sorted by absolute correlation):")
display(quality_correlation.abs().sort_values(ascending=False))
```
